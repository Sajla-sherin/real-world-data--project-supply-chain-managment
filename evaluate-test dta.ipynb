# missing valuesÂ¶

```python
df.isnull().sum()*100/len(df)
```

```python
#we can drop the warehouse estimated feature as there is 47% missing data and also because there is no need to predict the dependent variable

df.drop(['wh_est_year'],axis=1,inplace=True) 
```

```python
#The missing values in"approved_wh_govt_certificate" are low,  so we can use mode value fill as it is a categorical variable
df.approved_wh_govt_certificate.fillna(df.approved_wh_govt_certificate.mode()[0],inplace=True)
```

```python
#missing values are low"workers_num" , so we can use mean value fill as it is numerical variable
df.workers_num.fillna(df.workers_num.mean(),inplace=True)
```

```python
#now the missing values completely eradicated
df.isnull().sum()*100/len(df)

```

```python
# here 'WH_regional_zone' converted into numerical type by slicing the zone in it
df['WH_regional_zone' ]=df['WH_regional_zone'].apply(lambda x: x[-1])
df.WH_regional_zone=pd.to_numeric(df.WH_regional_zone)
```

#### all categorical variables are converted into numerical type using one hote

```python
# List of categorical columns to one-hot encode
ob_list = ['Location_type', 'WH_capacity_size', 'zone', 'wh_owner_type', 'approved_wh_govt_certificate']

# Convert the specified columns to 'category' data type
df[ob_list] = df[ob_list].astype('category')

# Perform one-hot encoding with drop_first=True
df1 = pd.get_dummies(df, columns=ob_list, prefix=ob_list, drop_first=True)

# Convert the one-hot encoded columns to integers (0 or 1)
df1 = df1.astype(int)

# Display the encoded DataFrame
print(df1.head())
```

## splitting of dependent variables and independent variable

```python
x = df1.drop(['product_wg_ton'],axis = 1)
```

```python
x.head()
```

```python
y = df1['product_wg_ton']
```

```python
y.head()
```

## splitting data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 101)
```

## standardaization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
y_train
```

### to evaluate the "test data" set using "RandomForestRegressor"

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize and train the Random Forest Regressor model
random_forest = RandomForestRegressor(random_state=42)  # You can add hyperparameters if needed
random_forest.fit(x_train, y_train)

# Make predictions on the training and test datasets
y_train_pred_rf = random_forest.predict(x_train)
y_test_pred_rf = random_forest.predict(x_test)

# Evaluate the model on the training dataset
train_mae_rf = mean_absolute_error(y_train, y_train_pred_rf)
train_rmse_rf = mean_squared_error(y_train, y_train_pred_rf, squared=False)
train_r2_rf = r2_score(y_train, y_train_pred_rf)

print('Random Forest Regressor Model performance for Training set:')
print(f"- Mean Absolute Error (MAE): {train_mae_rf:.4f}")
print(f"- Root Mean Squared Error (RMSE): {train_rmse_rf:.4f}")
print(f"- R-squared (R2) Score: {train_r2_rf:.4f}")

# Evaluate the model on the test dataset
test_mae_rf = mean_absolute_error(y_test, y_test_pred_rf)
test_rmse_rf = mean_squared_error(y_test, y_test_pred_rf, squared=False)
test_r2_rf = r2_score(y_test, y_test_pred_rf)

print('\nRandom Forest Regressor Model performance for Test set:')
print(f"- Mean Absolute Error (MAE): {test_mae_rf:.4f}")
print(f"- Root Mean Squared Error (RMSE): {test_rmse_rf:.4f}")
print(f"- R-squared (R2) Score: {test_r2_rf:.4f}")
```

```python
plt.scatter(y_test,y_test_pred_rf);
plt.xlabel('Actual');
plt.ylabel('Predicted');
```

```python

# Plot predicted vs. actual values for the test set
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_test_pred_rf, alpha=0.5)
plt.xlabel("Actual Product Weight")
plt.ylabel("Predicted Product Weight")
plt.title("Random Forest Regressor: Predicted vs. Actual")
plt.grid(True)
plt.show()
```
